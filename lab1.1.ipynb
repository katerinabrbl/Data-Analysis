```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#generating data
size = 50
a = -0.9
b = 3.3
x = np.random.random(size=size)
noise = np.random.random(size=size)

y = a*x+b+noise

#setting metrics space and view
f = plt.figure(figsize=(10,10))
c = f.add_subplot(111)
c.scatter(x, y, marker='.')
c.set_aspect('equal')

#Regression
data = {'X': x, 'Y': y}
df = pd.DataFrame(data, columns = ['X','Y'])
X = df['X'].values
Y = df['Y'].values

#mean of x and y
mean_x = np.mean(X)
mean_y = np.mean(Y)

data = df

#predict
m = len(X)
numer = 0
denom = 0
for i in range(m):
  numer += (X[i] - mean_x) * (Y[i] - mean_y)
  denom += (X[i] - mean_x) ** 2
m = numer / denom
c = mean_y - (m * mean_x)

print (f'a_predicted = {m} \na = {a} \nb_predicted = {c} \nb = {b}')

max_x = np.max(X)+1
min_x = np.min(Y)-4

x = np.linspace (min_x, max_x, 5)
y = c + m * x

#plotting the regression line
plt.plot(x, y, c='purple', label='Regression Line')
#plotting the data points
plt.scatter(X, Y, c='pink', label='data points')
#putting labels for x and y axis
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
