```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

import seaborn as sns
```

```python
dfhouse = pd.read_csv("D:\Универ\ParisHousingClass.csv",sep=",")
```

```python
continuousColumns = ["squareMeters","numberOfRooms","hasPool", "floors", "cityCode", "cityPartRange", "numPrevOwners", "made", "isNewBuilt", "hasStormProtector", "basement", "attic" ,"garage" ,"hasStorageRoom" ,"hasGuestRoom","price"]
willOneHotEncode = ["category"]
willScale  = continuousColumns
```

```python
for item in willOneHotEncode:
    dfhouse = pd.concat([dfhouse,pd.get_dummies(dfhouse[item],prefix=item)],axis=1)
    dfhouse = dfhouse.drop(columns=[item])
```

```python
minMaxScaler = MinMaxScaler()
scaledColums = pd.DataFrame(minMaxScaler.fit_transform(dfhouse[willScale]),columns=willScale)
```

```python
scaler = StandardScaler()
X_std = scaler.fit_transform(dfhouse)
pca = PCA()
X_pca = pca.fit_transform(X_std)
exp_var_pca = pca.explained_variance_ratio_
cum_sum_eigenvalues = np.cumsum(exp_var_pca)
```

```python
plt.bar(range(0,len(exp_var_pca)), exp_var_pca, align='center', label='Explained variance')
plt.step(range(0,len(cum_sum_eigenvalues)), cum_sum_eigenvalues, where='mid',label='Cumulative explained variance')
plt.ylabel('Explained variance')
plt.xlabel('Principal component')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
pca = PCA(n_components=2)
pca_result = pca.fit_transform(dfhouse)
```

```python
pca_result_df = pd.DataFrame({'pca_1': pca_result[:,0], 'pca_2': pca_result[:,1]})

fig, ax = plt.subplots(1)
sns.scatterplot(x='pca_1', y='pca_2', data=pca_result_df, ax=ax,s=10)
lim = (pca_result.min()-5, pca_result.max()+5)
ax.set_xlim(lim)
ax.set_ylim(lim)
ax.set_aspect('equal')
```

```python
tsne = TSNE(n_components=2)
tsne_result = tsne.fit_transform(dfhouse)
```

```python
tsne_result_df = pd.DataFrame({'tsne_1': tsne_result[:,0], 'tsne_2': tsne_result[:,1]})
fig, ax = plt.subplots(1)
sns.scatterplot(x='tsne_1', y='tsne_2', data=tsne_result_df, ax=ax,s=10)
lim = (tsne_result.min()-5, tsne_result.max()+5)
ax.set_xlim(lim)
ax.set_ylim(lim)
ax.set_aspect('equal')
```
