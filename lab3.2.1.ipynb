```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin
from sklearn.utils import shuffle
from PIL import Image
```

```python
n_colors = 64
sky = Image.open("D:\Универ\dat.JPG")
sky = np.array(sky, dtype=np.float64)/255
```

```python
w, h, d = original_shape = tuple(sky.shape)
assert d == 3
image_array = np.reshape(sky, (w * h, d))
```

```python
image_array_sample = shuffle(image_array, random_state=0, n_samples=1_000)
kmeans = KMeans(n_clusters=n_colors, random_state=0).fit(image_array_sample)
labels = kmeans.predict(image_array)
```

```python
def re_image(codebook, labels, w, h):
    return codebook[labels].reshape(w, h, -1)
```

```python
plt.figure(1, figsize=(10,6))
plt.clf()
plt.axis('off')
plt.title('Primary')
plt.imshow(sky)

plt.figure(2, figsize=(10,6))
plt.clf()
plt.axis('off')
plt.title(f'Quantized image ({n_colors} colors, K-Means)')
plt.imshow(re_image(kmeans.cluster_centers_, labels, w, h))
```
